```python
def log_cal(x):
    if x == 0:
        return np.log(0.001)
    else:
        return np.log(x)
```

```python
def new_features(dataframe):
    #dataframe['delta'] = dataframe.delta + 0.001
    dataframe['Local_Re'] = dataframe["VELOC:0"] * dataframe.delta / dataframe.viscosity
    #dataframe['Local_Re'] = dataframe.Local_Re + 0.001
    dataframe['Local_Re_Avg'] = dataframe["AVVEL:0"] * dataframe.delta / dataframe.viscosity
    dataframe['Local_Re_y'] = dataframe["VELOC:0"] * dataframe["Points:1"] / dataframe.viscosity
    dataframe['delta_squared'] = np.square(dataframe.delta)
    dataframe['Local_Re_log'] = dataframe.Local_Re.apply(log_cal)
    dataframe['Log_delta'] = dataframe.delta.apply(log_cal)
    dataframe['y_delta'] = dataframe.delta / dataframe['Points:1']
```
