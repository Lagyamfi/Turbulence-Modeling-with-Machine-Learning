```python
#load test data
test_1_data = pd.read_csv("./Experiments/exp_1/test_data/test_1/test_1.csv")
```

```python
# plot evolution of training and validation results
results = test_model.evals_result()
epochs = len(results['validation_0']['rmse'])
x_axis = range(0, epochs)

# plot RMSE
fig, ax = plt.subplots()
ax.plot(x_axis, results['validation_0']['rmse'], label='Train', c="b")
ax.plot(x_axis, results['validation_1']['rmse'], label='Validation', c="r")
ax.set(title=" Train and Validation RMSE")
ax.set_xlabel("Iteration")
ax.set_ylabel("RMSE")
ax.legend()
```

```python
#create model object
#test_1_data = test_1_data.drop(columns=['delta'])
test_1 = XGB_test(test_model, test_1_data)
```

```python
test_1.plot_importance()
```

```python
test_1.compare_values(plot=True)
```

```python
test_1.score("r2", plot_residuals=True)
```

```python
test_1.make_plot(title="Train on re_tau = 1000, Test on re_tau=180")
```
